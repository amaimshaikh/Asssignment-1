```python
def reverseName(name):
    revName = ""
    for i in name:
        revName = i + revName
    return (revName)

firstName = str(input("Enter first name: "))
lastName = str(input("Enter last name: "))
fullName = firstName + " " + lastName
print ("Reversed name: " + reverseName(fullName))
```
